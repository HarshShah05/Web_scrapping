Beautify data

```python
BeautyData=BeautifulSoup(Htmlpage,"html.parser")
```

```python
BeautyData
```

Get all products boxes

Going into 1 link html page

```python
oplink=requests.get(pagelink)
oplink
```

```python
opblink=BeautifulSoup(oplink.text,"html.parser")
opblink
```

```python
# opblink

custdet=opblink.find_all("div",{"class":"a-section review aok-relative"})
custdet
```

```python
len(custdet)
```

Names

```python
name=custdet[0].find_all("div",{"class":"a-profile-content"})[0].text
name
```

```python
names=[]

for i in custdet:
  names.append(i.find_all("div",{"class":"a-profile-content"})[0].text)

for j in names:
  print(j)
```
