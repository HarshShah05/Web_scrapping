```python
import requests 
from bs4 import BeautifulSoup 
from urllib.request import urlopen
import logging
```

Going to flipkart link

```python
# scrapping data of iphone 11 reviews
flipcart_url="https://www.flipkart.com/search?q="+"tv"
```

```python
flipcart_url
```

Opening url with urlopen


```python
urlclient=urlopen(flipcart_url)
```

```python
urlclient
```

Get html of page

```python
# we get entire html of page
flipcart_page=urlclient.read()
```

```python
flipcart_page
```

Html in beuatified form

```python
bhtml=BeautifulSoup(flipcart_page,'html.parser')
```

```python
bhtml
```
